```python
def getNthFib(n):
    num1, num2 = 0, 1
    if n == 0:
        raise Exception("n cannot be 0")
    if n == 1:
        return num1
    for i in range(3, n+1):
        temp = num2
        num2 = num1 + num2
        num1 = temp
    return num2
```

```python
def track_river(matrix, visited, h, w, H, W):
    if h >= H or h < 0 or w >= W or w < 0:
        return 0
    size = 0
    if  matrix[h][w] == 1 and visited[h][w] == 0:
        visited[h][w] = 1
        size += 1
        size += track_river(matrix, visited, h, w+1, H, W)
        size += track_river(matrix, visited, h+1, w, H, W)
        size += track_river(matrix, visited, h, w-1, H, W)
        size += track_river(matrix, visited, h-1, w, H, W)
    return size

def riverSizes(matrix):
    sizes = []
    visited = [[0 for col in row] for row in matrix]
    H = len(matrix)
    W = len(matrix[0])
    for h in range(H):
        for w in range(W):
            riv_size = track_river(matrix, visited, h, w, H, W)
            if riv_size != 0:
                sizes.append(riv_size)
    return sizes
```

```python
def add_to_existing(n, results):
    for r in results:
        s = r[0]
        e = r[1]
        if n in range(s,e+1):
            return True
        if n == s-1:
            r[0] = n
            return True
        if n == e+1:
            r[1] = n
            return True
    return False
    
def func(i, r, results, visited, new_results):
    for j in range(i+1, len(results)):
        if visited[j] == 1:
            continue
        if r[1] + 1 == results[j][0]:
            new_results.append([r[0], results[j][1]])
            visited[j] = 1
            return
        elif r[0] - 1 == results[j][1]:
            new_results.append([results[j][0], r[1]])
            visited[j] = 1
            return
    new_results.append(r)
    return

def concat_arrays(results):
    visited = [0 for i in results]
    new_results = []
    for i, r in enumerate(results):
        if visited[i] == 0:
            func(i, r, results, visited, new_results)
    return new_results
    
def largest_range(array):
    results = []
    for n in array:
        added = add_to_existing(n, results)
        if not added: 
            results.append([n,n])
    while(True):
        new_results = concat_arrays(results)
#         print(f"results: {results} | new results: {new_results}")
        if new_results == results:
            break
        results = new_results

    sorted_results = sorted(results, key=lambda x: x[1]-x[0])
    
    return sorted_results[-1]
```

```python
def largestRange(array):
    seen = {x:0 for x in array}
    left = right = 0
    for num in array:
        if seen[num] == 0:
            l = num - 1
            r = num + 1
            
            while l in seen:
                seen[l] = 1
                l -= 1
            l += 1
            
            while r in seen:
                seen[r] = 1
                r += 1
            r -= 1
            
            if (right - left) <= (r - l):
                left = l
                right = r
    return [left, right]
```

```python
# the sort operation leaves this as O(nlog(n)), which is bad
# O(1) in space
def twoNumberSum(array, targetSum):
    array.sort()
    l, r = 0, len(array) - 1
    while l < r:
        s = array[l] + array[r]
        if s == targetSum:
            return [array[l], array[r]]
        elif s < targetSum:
            l += 1
        elif s > targetSum:
            r -= 1
    return []
```

```python
# O(n) in time, and O(n) in space
def twoNumberSum(array, targetSum):
    seen = {x:1 for x in array}
    for n in array:
        seen[n] = 0
        missing = targetSum - n
        if missing in seen and seen[missing]:
            return [n, missing]
    return []
```

```python
def twoNumberSum(array, targetSum):
    nums = {}
    results = []
    for num in array:
        missing = targetSum - num
        if missing in nums:
            results.append([num, missing])
        else:
            nums[num] = "bla"
    return results

def threeNumberSum(array, targetSum):
    results = []
    for i, num in enumerate(array):
        newTarget = targetSum - num            
        newArray = [*array[:i], *array[i+1:]]
        twoNums = twoNumberSum(newArray, newTarget)
        if twoNums:
            results.append([num, *twoNums])
    return results
```

```python
def threeNumberSum(array, targetSum):
    results = []
    array.sort()
    for i, n in enumerate(array):
        l = i+1
        r = len(array) - 1
        diff = targetSum - n
        while l < r:
            s = array[l] + array[r]
            if s == diff:
                results.append([n, array[l], array[r]])
                l += 1
                r -= 1
            elif s < diff:
                l += 1
            elif s > diff:
                r -= 1
    return results
```

```python
def bubbleSort(array):
    isSwap = True
    while isSwap:
        isSwap = False
        for i in range(len(array)-1):
            if array[i] > array[i+1]:
                tmp = array[i]
                array[i] = array[i+1]
                array[i+1] = tmp
                isSwap = True
    return array
```

```python
def swap(array, i, j):
    tmp = array[i]
    array[i] = array[j]
    array[j] = tmp

def quickSort(array):
    if not array:
        return array
    
    if len(array) == 1:
        return array

    pivot = array[0]
    left = 1
    right = len(array) - 1
    while right >= left:
        if (array[left] > pivot) and (array[right] < pivot):
            swap(array, left, right)
        if array[left] <= pivot:
            left += 1
        if array[right] >= pivot:
            right -= 1
    swap(array, 0, right)
    array[:right] = quickSort(array[:right])  # these operations are wasteful in space
    array[right+1:] = quickSort(array[right+1:])  # these operations are wasteful in space
    return array
```

```python
quickSort([3, 1,2])
```

```python
def swap(array, i, j):
    array[i], array[j] = array[j], array[i]

def quickSortHelper(array, start, end):
    if end - start < 1:
        return 
    
    pivot = array[start]
    left = start + 1
    right = end
    while right >= left:
        if (array[left] > pivot) and (array[right] < pivot):
            swap(array, left, right)
        if array[left] <= pivot:
            left += 1
        if array[right] >= pivot:
            right -= 1
    swap(array, start, right)
    quickSortHelper(array, start, right-1) 
    quickSortHelper(array, right+1, end) 
    return array

def quickSort(array):
    quickSortHelper(array, 0, len(array)-1)
    return array
```

```python
import math

# O(n^2)
def smallestDifference(arrayOne, arrayTwo):
    minDistance = math.inf
    result = []
    for a1 in arrayOne:
        for a2 in arrayTwo:
            if abs(a1-a2) < minDistance:
                result = [a1, a2]
                minDistance = abs(a1-a2)
    return result
```

```python
def smallestDifference(arrayOne, arrayTwo):
    minDistance = math.inf
    result = []
    arrayOne.sort()
    arrayTwo.sort()
    i = j = 0
    while i < len(arrayOne) and j < len(arrayTwo):
        if abs(arrayOne[i]-arrayTwo[j]) < minDistance:
            result = [arrayOne[i], arrayTwo[j]]
            minDistance = abs(arrayOne[i]-arrayTwo[j])
        if arrayOne[i] < arrayTwo[j]:
            i += 1
        else:
            j += 1
    return result
```

```python
def skyline(heights):
    diffs = [e1 - e2 for e1, e2 in zip(heights, heights[1:])]
    return sum(diff for diff in diffs if diff > 0) + heights[-1]
```

```python
def solution(N):
    array = str(N)
    result = []
    mod = 1
    for i in range(len(array) + 1):
        if i == len(array):
            result.append('5')
            break
        
        if array[i] == '-':
            mod = -1
            result.append(array[i])
            continue
        
        if int(array[i]) * mod > 5 * mod:
            result.append(array[i])
        else:
            result.append('5')
            result.append(array[i:])
            break
        
    return int("".join(result))
```

```python
def solution(n: int, target_num: int = 5):
    target_digit = str(target_num)
    digits = str(n)

    mod, result, checked_digits = (1, [], digits) if n >= 0 else (-1, ['-'], digits[1:])

    for i, digit in enumerate(checked_digits):
        if int(digit) * mod > target_num * mod:
            result.append(digit)

        else:
            result.append(target_digit)
            result.append(checked_digits[i:])
            break

    else:
        result.append(target_digit)

    return int("".join(result))
```
